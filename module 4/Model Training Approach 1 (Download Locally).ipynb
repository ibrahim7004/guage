```python
import PyPDF2
from langchain.prompts import PromptTemplate
from langchain.llms import CTransformers
```

```python
path = r'C:\Users\hp\Desktop\preply\lars\analyst reports\test\EQUITY Factsheet_ Adobe Inc. (1).pdf'
```

```python
def read_document(file_path):
    with open(file_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ''
        for page_num in range(len(reader.pages)):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
document_content = read_document(path)
```

```python
model_path2 = r"C:\Users\hp\Desktop\sem6\PL Lab\lab10\llama-2-13b.ggmlv3.q2_K.bin"
```

    Temperature:
- 'temperature' parameter controls the randomness of the model's generated reponse. Ranges from 0-1
- Higher temp introduces more randomness/creativity in the reponses (closer to 1)
- Lower temp means less randomness (closer to 0)
- In essence, temperature acts like a dial between creativity and predictability.

    model_type="llama"
- This parameter is responsible for specifying the type of LLM being used (Llama in this case)
- Since we are interacting with a llama model, as specified by our model_path variable, we must inform the model about it.

    Max New Tokens
- This parameter sets the max numbe of new tokens the LLM can generate in reposnse to a prompt.
- A token refers to a basic unit of text processed by aa model, which may be a word, punctuation mark, or special character.
- 256 value ensures that the LLM will restrict its reponse to 256 tokens.
- Why might we want to set a max value for this?
    - Efficiency (time improvement, resources)
    - Focus of the model on the core idea instead of getting distracted -> hallucination.
    - Relevance of the response.

```python
def get_llama_response(prompt, temperature=0.7):
    llm = CTransformers(model=model_path2,
                        model_type="llama",
                        config={"max_new_tokens": 256, "temperature": temperature})
    response = llm(prompt)
    return response
```

```python
while True:
    user_input = input("Give me a 20 word asnwer of the passed text, do not exceed 20 words. Use the data provided only, not your own knowledge.:")
    
    if user_input.lower() in ['q', 'quit']:
        print("Exiting")
        break
    
    max_tokens = 50  
    chunks = [document_content[i:i + max_tokens] for i in range(0, len(document_content), max_tokens)]
    
    for chunk in chunks:
        response = get_llama_response(chunk)
        print(response)
```
